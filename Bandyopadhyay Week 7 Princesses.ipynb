# 12 Dancing Princesses

```python
#import NLTK things
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from string import punctuation
```

```python
#open file and read it
filepath = '12dancingprincesses.txt'
princess = open(filepath, "r", encoding = 'utf8', errors="ignore")

text = princess.read()
print(text)
```

## Tokenizing Words

```python
#tokenize me, captain!
word_tokenize(text)
```

```python
#convert to lowercase for consistency and then count all the words
text2 = text.lower()

tknz = word_tokenize(text2)
len(tknz)
```

## Finding Most Common Words

```python
#get a word count, essentially all words and puntuation, ordered by usage, descending
wct = FreqDist(tknz)
wct
```

## Removing Punctuation and Stopwords

```python
#removing punctuation from the original list of tokens, because punctuation aren't words
for token in tknz:
    if token in punctuation:
        tknz.remove(token)
```

```python
#count all the actual words
len(tknz)
```

```python
#a lot of the words in the word count are really common grammatical words, called stopwords here
eng_stopwords = stopwords.words('english')
eng_stopwords
```

```python
#count for stopwords
rm_count = 0

#list for new words
new_words = []

#put non-stopwords in their own list and just count how many stopwords there are
for token in tknz:
    if token not in eng_stopwords:
        new_words.append(token)
    else: rm_count += 1
```

```python
#this many stopwords
rm_count
```

```python
#this many non-stopwords
len(new_words)
```

## Top Ten Words in this Story

```python
#make a word count for non-stopwords, and show the top ten, and you can get a good idea of the content of the story
wct_nw = FreqDist(new_words)
wct_nw.most_common(10)
```
