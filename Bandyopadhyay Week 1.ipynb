```python
#setting up pandas, sql, and viewing options
import pandas as pd

pd.set_option('max_rows', 81)

from sqlalchemy import create_engine, inspect
```

```python
#identifying the database file and creating the engine
db_file = r'boardgames.sqlite'

engine = create_engine(f"sqlite:///{db_file}")
engine.table_names()
```

### BoardGames Table

```python
#Shows information about each of the columns
inspector = inspect(engine)
boardgame_fields = inspector.get_columns('BoardGames')
boardgame_field_names = pd.DataFrame.from_dict(boardgame_fields)
boardgame_field_names
```

```python
#shows the actual table
sql = "SELECT * FROM BoardGames;"
boardgames_df = pd.read_sql(sql, engine)
boardgames_df
```

```python
boardgames_df.shape
```
